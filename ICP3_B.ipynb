```python
# Try to plot the 'Price' column using the plot() function
car_sales['Price'].plot(title='Price Column Plot')
plt.show()
```

Why doesn't it work?

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
cleaned = car_sales['Price'].astype(str).str.replace('[\$,]', '', regex=True)

cleaned = cleaned.str[:-3]

car_sales['Price_cleaned'] = pd.to_numeric(cleaned, errors='coerce')

print(car_sales[['Price', 'Price_cleaned']].head())
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column


# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)

# View the car_sales DataFrame
from datetime import datetime, timedelta

car_sales['Total Sales'] = car_sales['Price_cleaned'].cumsum()

today = datetime.today()

dates = [today + timedelta(days=i) for i in range(len(car_sales))]

car_sales['Sale Date'] = dates

print(car_sales.head())
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
car_sales.plot(x='Sale Date', y='Total Sales', kind='line', title='Total Sales Over Time')
plt.xlabel('Sale Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)  # Rotate date labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Convert the 'Price' column to the integers

# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function
car_sales['Price_cleaned'] = (
    car_sales['Price']
    .astype(str)
    .str.replace('[\$,]', '', regex=True)
    .astype(float)
)

car_sales['Price_int'] = car_sales['Price_cleaned'].astype(int)

car_sales.plot.scatter(x='Odometer (KM)', y='Price_int', title='Odometer vs Price')

plt.xlabel('Odometer (KM)')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Create a bar graph of the 'Make' and 'Odometer (KM)' columns in the car_sales DataFrame
odometer_avg = car_sales.groupby('Make')['Odometer (KM)'].mean()

odometer_avg.plot(kind='bar', color='skyblue')
plt.title('Average Odometer by Car Make')
plt.xlabel('Make')
plt.ylabel('Average Odometer')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

Completing this step correctly should result in a figure which looks like plot3

```python
# Replot the same figure as above except change the "cmap" parameter
# of scatter() to "winter"
# Also change the "color" parameter of axhline() to "red"

# Create the plot


# Plot the data


# Customize the plot


# Add a meanline
sns.set_style('whitegrid')

over_50 = heart_disease[heart_disease['age'] > 50]

fig, ax = plt.subplots(figsize=(8,6))

scatter = ax.scatter(
    over_50['age'],
    over_50['chol'],    # changed here to 'chol'
    c=over_50['chol'],
    cmap='winter',
    edgecolor='k',
    alpha=0.7
)

fig.colorbar(scatter, ax=ax, label='Cholesterol')

mean_chol = over_50['chol'].mean()
ax.axhline(mean_chol, color='red', linestyle='dashed', linewidth=2)

ax.set_title('Age vs Cholesterol with Mean Line')
ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol')

plt.show()

axes = over_50.select_dtypes(include='number').hist(
    bins=30,
    layout=(4,4),
    figsize=(12, 30),
    edgecolor='black',
    alpha=0.7
)

for ax, col in zip(axes.flatten(), over_50.select_dtypes(include='number').columns):
    mean_val = over_50[col].mean()
    ax.axvline(mean_val, color='red', linestyle='dashed', linewidth=1)
    ax.set_title(f'{col} Distribution')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
